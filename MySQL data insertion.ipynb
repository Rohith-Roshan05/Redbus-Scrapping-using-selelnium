```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# csv to DataFrame
df_apsrtc = pd.read_csv("Z:\Redbus project\APSRTC (Andhra pradesh) bus details.csv")
df_ksrtc = pd.read_csv("Z:/Redbus project/KSRTC (Kerala) bus details.csv")
df_tsrtc = pd.read_csv("Z:/Redbus project/tsrtc (Telungana) bus details.csv")
df_ktcl = pd.read_csv("Z:/Redbus project/KTCL(goa) bus details.csv")
df_rsrtc = pd.read_csv("Z:/Redbus project/rsrtc(Rajasthan) bus details.csv")
df_sbstc = pd.read_csv("Z:/Redbus project/SBSTC(South Bengal) bus details.csv")
df_hrtc = pd.read_csv("Z:/Redbus project/HRTC(Haryana) bus details.csv")
df_upsrtc = pd.read_csv("Z:/Redbus project/UPSRTC(Uttar pradesh) bus details.csv")
df_pepsu = pd.read_csv("Z:/Redbus project/PEPSU(Punjab) bus details.csv")
df_bsrtc = pd.read_csv("Z:/Redbus project/BSRTC(Bihar) bus details.csv")

Final_df = pd.concat([df_apsrtc,df_tsrtc,df_ksrtc,df_ktcl,df_rsrtc,df_sbstc,df_hrtc,df_upsrtc,df_pepsu,df_bsrtc],ignore_index=True)
Final_df
```

```python
Final_df.info()
```

```python
# convert the price to numreic
Final_df["price"] = Final_df["price"].str.replace("INR", "").str.replace(",", "").str.strip()
Final_df["price"] = pd.to_numeric(Final_df["price"], errors='coerce')
Final_df["price"].fillna(0, inplace=True)
```

```python
if Final_df["star_rating"].dtype != "object":
    Final_df["star_rating"] = Final_df["star_rating"].astype(str)
Final_df["star_rating"] = Final_df["star_rating"].str.replace("New", "")
Final_df["star_rating"] = Final_df["star_rating"].str.strip()
Final_df["star_rating"] = Final_df["star_rating"].str.split().str[0]
Final_df["star_rating"] = pd.to_numeric(Final_df["star_rating"], errors='coerce')
Final_df["star_rating"].fillna(0, inplace=True)
```

```python
# info after the data type change
Final_df.info()
```

```python
# replace the nan value
Final_df = Final_df.replace(np.nan, None)
```

```python
Final_df['seat_availability'] = Final_df['seat_availability'].fillna('Unknown')
Final_df

```

```python
# Table creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS bus_details(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  bus_route_name VARCHAR(255) NOT NULL,
                  bus_route_link VARCHAR(255) NOT NULL,
                  bus_name VARCHAR(255) NOT NULL,
                  bus_type VARCHAR(255)NOT NULL,
                  departing_time VARCHAR(255)NOT NULL,
                  duration VARCHAR(255)NOT NULL,
                  reaching_time VARCHAR(255)NOT NULL,
                  star_rating VARCHAR(255)NOT NULL,
                  price FLOAT NULL,
                  seat_availability VARCHAR(255)NOT NULL
                  )''')
print("Table created successfully")
```
